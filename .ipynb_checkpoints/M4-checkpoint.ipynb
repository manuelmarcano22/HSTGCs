```python
import pandas as pd
from astropy.time import Time
import numpy as np
from tabulate import tabulate
```

```python
m4data = pd.read_csv('M4.csv',low_memory=False)
m4data = m4data.drop(m4data.index[0])
#m4data
```

```python
m4data['YearDay'] = m4data['Start Time'].map(lambda times: '.'.join(Time(times).yday.split(':',2)[:2]))

m4data['Week'] = m4data['Start Time'].map(lambda times: int(np.floor(float(Time(times).yday.split(':')[1])/7.)))
```

```python
t =m4data.groupby(['Filters/Gratings','Instrument','Exp Time','Proposal ID']).size().reset_index().rename(columns={0:'ocurrence'})
t.sort_values(by=['ocurrence'],ascending=False)
```

```python
print(tabulate(t.sort_values(by=['ocurrence'],ascending=False), tablefmt="pipe", headers="keys"))
```
